### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
cordinate1 = float(input("Enter  (x1): "))
cordinate2 = float(input("Enter (y1): "))
cordinate3 = float(input("Enter (x2): "))
cordinate4 = float(input("Enter (y2): "))

euclidean_distance = ((cordinate1 - cordinate3)**2 + (cordinate2 - cordinate4)**2)**0.5

print("Euclidean distance: ",euclidean_distance)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
# Write your code here
principle = float(input("Enter principle: "))
rate = float(input("Enter rate: "))
time = float(input("Enter time: "))

simple_interest = (principle * rate * time)/100

print("Simple interest: ",simple_interest)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
heads = int(input("Enter heads: "))
legs = int(input("Enter legs: "))

dogs = (legs -2*heads)/2
chicken = heads - dogs

print("dogs: ",dogs)
print("chicken: ",chicken)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n = int(input("Enter number: "))
sum = ( n *(n + 1)*(2*n+1))//6
print('sum of squares of n natural no: ', sum)

# 2nd loop method
sum = 0
for i in range(1, n+1):
    sum += i**2
print(sum)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
first_term = int(input("Enter first term: "))
second_term = int(input("Enter second term: "))
n = int(input("Enter n: "))

d = second_term - first_term
nth_term = first_term + (n-1)*d

print("Nth term of the series: ",nth_term)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
f1_num = int(input("Enter first fraction numerator: "))
f1_den = int(input("Enter first fraction denominator: "))
f2_num = int(input("Enter second fraction numerator: "))
f2_den = int(input("Enter second fraction denominator: "))

num = f1_num * f2_den + f2_num * f1_den
den = f1_den * f2_den

print("Sum of the fractions: ",num,"/",den)
````

5### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
import math

h_b1 = float(input('Enter the height of the tank: '))
l_b1 = float(input('Enter the length of the tank: '))
b_t1 = float(input('Enter the breadth of the tank: '))

g_h = float(input('Enter the height of the glass: '))
g_r = float(input('Enter the radius of the glass: '))

vol_tank = (h_b1 * l_b1 * b_t1)
vol_glass = 3.14 * (g_r**2) *g_h

print('no of glasses: ', math.floor(vol_tank/vol_glass))
````
