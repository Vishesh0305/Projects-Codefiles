Imbalanced Dataset:
A dataset with an unequal class distribution

```python

# importung the dependencies
import numpy as np
import pandas as pd
```

```python
# loading the dataset to pandas DataFrame
credit_Card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# first five rows of the dataframe
credit_Card_data.head()
```

```python
credit_Card_data.tail()
```

```python
# distribution of the two classes
credit_Card_data['Class'].value_counts()
```

This is Highly imbalanced dataset

0 -->  Legit Transaction and 1 --> Fraudulent Transaction

```python
# Separating the legit and fraudulent transactions
legit = credit_Card_data[credit_Card_data.Class == 0]
fraud = credit_Card_data[credit_Card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

 Under Sampling

 Build a sample dataset containing similar distribution of Legit & Fraudulent Transactions

 No of Fraudulent Transactions --> 3

```python
legit_sample = legit.sample(n = 103)
```

```python
print(legit_sample.shape)
```

Concatenate the two dataframes


```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
