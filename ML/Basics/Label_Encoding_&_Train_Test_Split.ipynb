```python
# Importing the Dependencies
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
# Loading the data from csv to Pandas dataframe
cancer_data = pd.read_csv('/content/breast_cancer_data.csv')
#first five rows of dataframe
cancer_data.head()
```

```python
#finding the count of diffrent labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the label encoder function
l_encode = LabelEncoder()
```

```python
labels = l_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataframe
cancer_data['target'] = labels
cancer_data.head()
```

In the target column 0 --> Benign , 1 --> Malignant

```python
cancer_data['target'].value_counts()
```
