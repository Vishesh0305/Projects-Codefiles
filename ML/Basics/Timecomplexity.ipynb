# ORDERS OF GROWTH
Goals:
- want to evaluate program's efficiency when input is very big
- want to express the growth of program's run time as input size grows
- want to put an upper bound on growth - as tight as possible
- do not need to be precise : "order of" not "exact" growth
- we will look at largest factors in run time (which section of the program will take the longest to run?)
- thus , generally we want tight upper bound on growth , as function of size of input, in worst case


```````````python
# factorial compute
def fact_iter(n):
  answer = 1  #1 operation
  while n >1:   #  Inside +1 op everytime condition checks
    answer *= n # 2 operations ( multiply and assign)
    n -= 1 # 2 operations ( subtract and assign)
  return answer # 1 operation
    # loop runs n-1 times with 5 ops every time = 5(n-1)
  # Total no of operations =1 + 5(n-1) + 1 + 1 = 5n - 2
  # But we ignore all constants and all multiplicative constants as well
  # So the time Complexity will be (O)N
```````````

#Types of ORDERS OF GROWTH

#Constant Time Complexity - A function has **constant time complexity** if it takes same amount of time reagrdless of input size.
- It does not depend on n
- Time stays fixed
- Graph is a horizontal line
- Notation is O(1)

```````````python
n = list(range(1, 11))

# Constant time (e.g., always 5 operations)
time = [5 for _ in n]

plt.plot(n, time)
plt.xlabel('Input size (n)')
plt.ylabel('Time (operations)')
plt.title('Constant Time Complexity O(1)')
plt.show()
```````````

# Linear Time Complexity - A function has **linear time complexity** if its running time increases  directly proportional to  size of the input (n).
- If n doubles --> time also doubles
- Graph is a slanted line.
- Notation is O(n).

Ex - Traversing through an array

```````````python
import numpy as np
n = np.arange(0, 100, 1)

T = n

plt.plot(n, T)
plt.xlabel("Input Size (n)")
plt.ylabel("Time Taken T(n)")
plt.title("Linear Time Complexity O(n)")

plt.grid()

plt.show()
```````````

#Quadratic Time Complexity - A function has **quadratic time complexity** when its running time grows proportional to the square of the input size.
- If n doubles --> time becomes 4 times, If n triples --> time becomes 9 times
- Graph = curved (parabolic) ,gets steep quick
- Notation is O(n)².

Ex- Nested loops

```````````python
#  Quadratic Time complexity
n = np.arange(0,50,1)
T = n**2
plt.plot(n,T)
plt.xlabel("Input Size (n)")
plt.ylabel("Time Taken T(n)")
plt.title("Quadratic Time Complexity O(n²)")
plt.grid()
plt.show()
```````````

#Logarithmic Time Complexity - A function has **logarithmic time complexity** where running time grows increases very slowly as the input size grows.
- If input doubles --> runtime increases ny only a small constant.
- Graph grows rapidly at start and flattens out quickly.  
-  Notation is O(log n).

Ex - Binary Search


```````````python
n = np.linspace(1,100,100)
log_n = np.log2(n)

plt.figure()
plt.plot(n,log_n)
plt.xlabel("Input Size (n)")
plt.ylabel("Time Taken T(n)")
plt.title("Logarithmic Time Complexity O(log n)")
plt.grid()
plt.show()
```````````

#Linearithmic Time Complexity - A function has linear time complexity whose running time grows proportionally to n log n ( a combination of linear growth and logarithmic growth.)
- Slower than O(n) and Faster than O(n²) efficient for large datasets
- Notation is O(n log n)[linear * logarithmic growth]
- Graph rises rapidly at start, then flattens quickly

Ez - Merge Sort, Quick Sort,Heap Sort

```````````python
# Linearithmic Time Complexity
n = np.linspace(1,100,500)
linearithmic = n * np.log2(n)

plt.plot(n , linearithmic,linewidth =2)
plt.xlabel("Input Size (n)")
plt.ylabel("Time Taken T(n)")
plt.title("Linearithmic Time Complexity O(n log n)")
plt.grid(alpha=0.3)
plt.show()
```````````

# Exponential Time Complexity - A function has **exponential time complexity** when the running time grows proportionally to an exponential function of the input size.
- Very slow for large inputs and not practical for large datasets
- Doubles (or  worse) with each increment in input
- Starts off slow , then shoots up dramatically and becomes almost vertical
- Notation is O(2ⁿ)

Ex - used in brute force recursion

```````````python
# Exponential time complexity
n = np.linspace(1,10,100)
exponential = 2**n

plt.plot(n,exponential)
plt.xlabel("Input Size (n)")
plt.ylabel("Time Taken T(n)")
plt.title("Exponential Time Complexity")
plt.grid(alpha=0.3)
plt.show()
```````````

# Law of Addition
 Law of Addition for O():
 - used with sequential statements
 - O(f(n)) + O (g(n)) is o( f(n) + g(n) )

 Ex :
 ```
 for i in range(n):      
   print('a')            
 for j in range(n*n):    
   print('b')            
 ```
is O(n) + o(n*n) = O( n + n²) because of dominant term

# Law of Multiplication

Law of Multipliication for O():
- used with nested statements /loops
- O(f(n)) * O (g(n)) is O( f(n) * g(n) )

Ex :
````
for i in range(n):
    for j in range(n):
        print('a')
````
is O(n) * O(n) = O (n*n) = O(n²) because the outer loop goes n times and the inner loop goes n times for every outer loop iter.
