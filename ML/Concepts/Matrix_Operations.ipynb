```
Matrix with all values as 1
```

`````python
matrix_3 = np.ones((4,4))
print(matrix_3)
# float point values
`````

`````python
# integer values
matrix_3 = np.ones((4,4),dtype=int)
print(matrix_3)
`````

```
 Null Matrix --> Matrix with all values as 0
```

`````python
# integer values
null_matrix = np.zeros((4,4),dtype=int)
print(null_matrix)
`````

`````python
# floating point values
null_matrix = np.zeros((4,4))
print(null_matrix)
`````

```
Identity Matrix --> All diagonal values are 1
```

`````python
# integer values
identity_matrix = np.eye(5,5, dtype=int)
print(identity_matrix)
`````

`````python
# float point values
identity_matrix = np.eye(5,5)
print(identity_matrix)
`````

```
Matrix Addition --> Two Matrices can be added only if they have the same shape
```

`````python
# creating two matrices
A = np.array([[2,3],[4,5]])
B = np.array([[5,3],[6,1]])
`````

`````python
print(A)
`````

`````python
print(B)
`````

`````python
A.shape
`````

`````python
B.shape
`````

`````python
sum = A + B
print(sum)
`````

`````python
# addition of  random int matrices of size 3 x 3

matrix_1 = np.random.randint(10, size =(3,3))
matrix_2 = np.random.randint(20, size=(3,3))
`````

`````python
print(matrix_1)
`````

`````python
print(matrix_2)
`````

`````python
sum = np.add(matrix_1,matrix_2)   ## or sum = matrix_1 + matrix_2
print(sum)
`````

```
Matrix subtraction --> Two Matrices can be subtracted only if they have the same shape
```

`````python
# subtraction of  random int matrices of size 3 x 3

matrix_1 = np.random.randint(10, size =(3,3))
matrix_2 = np.random.randint(20, size=(3,3))
`````

`````python
print(matrix_1)
`````

`````python
print(matrix_2)
`````

`````python
difference = np.subtract(matrix_2,matrix_1)
print(difference)
`````

```
Multiplying a Matrix by a Scalar
```


`````python
x = 5
y = np.random.randint(10,size=(4,4))
print(y)
`````

`````python
product = np.multiply(x,y)  # product = x * y
print(product)
`````

```
Multiplying 2 Matrices
```

`````python
matrix_1 = np.random.randint(5, size =(3,3))
matrix_2 = np.random.randint(5, size=(3,3))
`````

`````python
print(matrix_1)
`````

`````python
print(matrix_2)
`````

`````python
product = np.dot(matrix_1,matrix_2)
print(product)
`````

```
Element wise Matrix Multiplication -->Both Matrices should have same shape.
```

`````python
matrix_3 = np.random.randint(5, size =(3,3))
matrix_4 = np.random.randint(5, size=(3,3))
`````

`````python
print(matrix_3)
`````

`````python
print(matrix_4)
`````

`````python
product = np.multiply(matrix_3,matrix_4)
print(product)
`````
