`````````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
`````````

`````````python
sns.set()
`````````

Plotting a vector

`````````python
plt.quiver(0,0,4,5)
plt.show()
`````````

`````````python
plt.quiver(0,0,4,5, scale_units='xy',angles = 'xy', scale = 1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
`````````

`````````python
plt.quiver(0,0,4,5, scale_units='xy',angles = 'xy', scale = 1)
plt.quiver(0,0,-3,-6, scale_units='xy',angles = 'xy', scale = 1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
`````````

Addition of 2 vectors

`````````python
vector_1 = np.asarray([0,0,6,8])

vector_2 = np.asarray([0,0,-2,1])

sum = vector_1 + vector_2
print(sum)
`````````

`````````python
plt.quiver(0,0,6,8, scale_units='xy',angles = 'xy', scale = 1,color='b')
plt.quiver(0,0,-2,1, scale_units='xy',angles = 'xy', scale = 1, color='g')
plt.quiver(0,0,4,9, scale_units='xy',angles = 'xy', scale = 1, color='m')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
`````````

Subtraction of 2 vectors

`````````python
vector_1 = np.asarray([0,0,6,8])

vector_2 = np.asarray([0,0,-2,1])

differnce = vector_1 - vector_2
print(differnce)
`````````

`````````python
plt.quiver(0,0,6,8, scale_units='xy',angles = 'xy', scale = 1,color='b')
plt.quiver(0,0,-2,1, scale_units='xy',angles = 'xy', scale = 1, color='g')
plt.quiver(0,0,8,7, scale_units='xy',angles = 'xy', scale = 1, color='m')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
`````````

Multiplication of a vector by a scalar

`````````python
vector_1 = np.asarray([0,0,2,3])

vector_2 = 2 * vector_1

print(vector_2)
`````````

`````````python
plt.quiver(0,0,2,3, scale_units='xy',angles = 'xy', scale = 1, color='g')
plt.quiver(0,0,4,6, scale_units='xy',angles = 'xy', scale = 1, color='m')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
`````````

`````````python
vector_1 = np.asarray([0,0,2,3])

vector_2 = -0.5  * vector_1

print(vector_2)
`````````

`````````python
plt.quiver(0,0,2,3, scale_units='xy',angles = 'xy', scale = 1, color='g')
plt.quiver(0,0,-1,-1.5, scale_units='xy',angles = 'xy', scale = 1, color='m')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
`````````

```````
Dot Product of 2 vectors
```````

`````````python
a = np.array([2,3])

b = np.array([4,4])

a_dot_b = np.dot(a,b)
`````````

`````````python
print(a_dot_b)
`````````

The result we get during Dot Product is a Scalar rather than a vector.

`````````python
c = np.array([40,20,35])

d = np.array([20,10,15])

c_dot_d = np.dot(c,d)
`````````

`````````python
print(c_dot_d)
`````````

```````
Cross Product of 2 Vectors
````````

`````````python
a = np.array([2,3,0])

b = np.array([4,4,0])

a_cross_b = np.cross(a,b)
`````````

`````````python
print(a_cross_b)
`````````

`````````python
c = np.array([5,10,15])

d = np.array([8,51,4])

c_cross_d = np.cross(c,d)
`````````

`````````python
print(c_cross_d)
`````````

The result we get during Cross Product is Vector.

`````
Projection of "a" vector on "b" vector
`````

`````````python
a = np.array([2,5])

b = np.array([8,-6])

# magnitude of "v" vector
magnitude_of_b = np.sqrt(np.sum(b**2))

proj_of_a_on_b = (np.dot(a,b)/ magnitude_of_b**2)*b

print("Projection of a vector on b vector",proj_of_a_on_b)
`````````
